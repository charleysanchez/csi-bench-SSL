```python
from pathlib import Path
import scipy.io as sio

root = Path("../data/RawContinuousRecording")

records = []

for mat_file in root.rglob("*.mat"):
    user_env = mat_file.parts[-3]
    user, env = user_env.split("_")
    device = mat_file.parts[-2]

    mat = sio.loadmat(mat_file)
    csi = mat["csi_trace"]["csi"][0, 0]

    shape = csi.shape  # (N_tx, N_rx, N_subcarriers, N_samples)

    records.append({
        "user": user,
        "env": env,
        "device": device,
        "path": str(mat_file),
        "N_tx": shape[0],
        "N_rx": shape[1],
        "N_subcarriers": shape[2],
        "N_samples": shape[3]
    })

print(records[:3])
```

```python
import pandas as pd

df = pd.DataFrame(records)
df["antenna_config"] = list(zip(df["N_tx"], df["N_rx"]))

print(df.groupby("device")["antenna_config"].unique())
print(df.groupby("device")["N_subcarriers"].unique())
```

This is great. Each device has either 1 * 4 or 2 * 2 tx, rx pairs which means we have a single set size of 4 total pairings. With the 14 vs 58 subcarriers, we can try to interpolate and use the higher resolution (58)
